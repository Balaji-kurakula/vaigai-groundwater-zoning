# HydroSight TN - AHP Analysis

This notebook computes the AHP weights from the fixed pairwise matrix and generates the AHP groundwater potential zoning raster.

```python
from pathlib import Path
import sys

project_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd()
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src import AHP_PAIRWISE_MATRIX
from src.ahp import compute_ahp_weights, weighted_overlay
from src.visualize import plot_area_statistics

classified_dir = project_root / 'data' / 'classified'
output_dir = project_root / 'outputs'
output_dir.mkdir(parents=True, exist_ok=True)

classified_layers = {
    'geology': classified_dir / 'geology.tif',
    'geomorphology': classified_dir / 'geomorphology.tif',
    'lineament_density': classified_dir / 'lineament_density.tif',
    'soil': classified_dir / 'soil.tif',
    'lulc': classified_dir / 'lulc.tif',
    'slope': classified_dir / 'slope.tif',
    'rainfall': classified_dir / 'rainfall.tif',
    'drainage_density': classified_dir / 'drainage_density.tif',
    'twi': classified_dir / 'twi.tif',
    'ndvi': classified_dir / 'ndvi.tif',
}
weights = compute_ahp_weights(AHP_PAIRWISE_MATRIX)
weights
```
