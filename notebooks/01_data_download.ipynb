# HydroSight TN - Data Intake

This notebook validates the expected raw datasets, previews basic metadata, and prepares the project workspace. The workflow assumes datasets are manually placed in `data/raw/`.

```python
from pathlib import Path
import geopandas as gpd
import rasterio
import pandas as pd

project_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data_raw = project_root / 'data' / 'raw'
expected = [
    'slope.tif', 'dem.tif', 'smap_sm.h5', 'sentinel2_B8.tif', 'geology.shp',
    'geomorphology.shp', 'soil.shp', 'lulc.tif', 'rainfall.tif', 'ndvi.tif',
    'borewells.shp', 'ndvi_1km.tif', 'lst_1km.tif'
]
inventory = pd.DataFrame({'file': expected})
inventory['exists'] = inventory['file'].apply(lambda name: (data_raw / name).exists())
inventory
```

```python
with rasterio.open(data_raw / 'slope.tif') as src:
    print('Slope shape:', src.shape)
    print('Slope CRS:', src.crs)
    print('Slope bounds:', src.bounds)

geology = gpd.read_file(data_raw / 'geology.shp')
borewells = gpd.read_file(data_raw / 'borewells.shp')
print('Geology columns:', geology.columns.tolist())
print('Borewell columns:', borewells.columns.tolist())
borewells.head()
```
